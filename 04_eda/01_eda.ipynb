# EDA(Exploratory Data Analysis, 탐색적 데이터 분석)

- 데이터를 바로 모델에 넣기 전에, 데이터의 구조와 문제점, 패턴, 분석 가능성을 확인하는 과정
- 데이터의 품질을 평가하고 전처리 과정을 통해 모델에 적합한 형태로 데이터를 변환하는 단계도 포함된다.

수업용 한 문장:

> EDA는 데이터에게 "너는 어떻게 생겼고, 어디가 이상하고, 무엇을 말해줄 수 있니?"라고 질문하는 과정이다.

## EDA를 하는 이유

| 이유 | 확인 질문 |
|---|---|
| 데이터 구조 파악 | 행/열은 몇 개인가? 컬럼은 무엇을 의미하는가? |
| 데이터 품질 확인 | 결측치, 중복, 이상값이 있는가? |
| 분포 확인 | 값이 어디에 몰려 있고, 한쪽으로 치우쳐 있는가? |
| 관계 확인 | 타깃 또는 관심 변수와 관련 있어 보이는 변수가 있는가? |
| 전처리 방향 결정 | 제거, 대체, 변환, 파생변수 생성이 필요한가? |
| 인사이트 도출 | 데이터로 설명할 수 있는 사실은 무엇인가? |

EDA의 목적은 예쁜 그래프를 만드는 것이 아니라, **다음 분석/모델링 단계에서 어떤 판단을 해야 하는지 근거를 찾는 것**임.

## EDA와 전처리, 머신러닝의 관계

| 단계 | 역할 |
|---|---|
| EDA | 데이터를 이해하고 문제를 발견하는 진단 단계 |
| 전처리 | 결측치, 이상값, 자료형, 인코딩 등을 실제로 처리하는 단계 |
| 머신러닝 | 정리된 데이터로 규칙을 학습하고 예측하는 단계 |


> EDA는 데이터 분석의 마무리이면서 머신러닝의 출발점임.  
> EDA에서 찾은 문제와 패턴이 머신러닝의 피처 선택, 전처리, 평가 방향으로 이어짐.

## EDA 전체 순서

1. 데이터 수집 및 로드
2. 데이터 구조 및 변수 이해
3. 데이터 요약 및 기술 통계
4. 결측치, 중복, 이상값 탐지
5. 변수별 분포 시각화
6. 변수 간 관계 시각화
7. 상관관계 및 교차분석
8. 데이터 전처리 방향 정리
9. Feature Engineering 방향 정리
10. 최종 요약 및 인사이트 도출

중요한 점은 각 단계마다 **코드 실행 결과를 해석 문장으로 바꾸는 것**임.

## 01. 데이터 수집 및 로드

분석할 데이터를 불러오고, 데이터 출처와 목적을 확인함.

해야 할 일:

- 데이터 출처 확인
- 파일 형식 확인: CSV, Excel, DB, API 등

- 분석 목적 또는 타깃 변수 확인

- 이 데이터는 어디서 온 데이터인가?

- 수집 과정에서 누락되었을 가능성은 없는가?

확인 질문:

- 평균과 중앙값 차이가 큰가?



## 05. 변수별 분포 시각화

각 변수가 어떤 모양으로 분포하는지 확인.

자주 사용하는 그래프:

| 변수 유형 | 그래프 | 확인 내용 |
|---|---|---|
| 범주형 | countplot, bar chart | 범주별 개수 |
| 수치형 | histplot, histogram | 값이 몰린 구간 |
| 수치형 | boxplot | 중앙값, 사분위수, 이상값 후보 |

확인 질문:

- 값이 한쪽으로 치우쳐 있는가?
- 특정 범주에 데이터가 몰려 있는가?
- 극단적으로 큰 값이나 작은 값이 있는가?
- 로그 변환이나 구간화가 필요해 보이는가?


```python
import seaborn as sns
import matplotlib.pyplot as plt

# 범주형 변수 분포
sns.countplot(data=df, x="Sex")
plt.show()

# 수치형 변수 분포
sns.histplot(data=df, x="Age", bins=30, kde=True)
plt.show()
```



## 06. 변수 간 관계 시각화

두 개 이상의 변수가 어떤 관계를 갖는지 확.

자주 사용하는 그래프:

| 관계 | 그래프 | 예시 |
|---|---|---|
| 수치형 + 수치형 | scatterplot | 나이와 요금 관계 |
| 범주형 + 수치형 | boxplot, barplot | 성별에 따른 요금 차이 |
| 범주형 + 타깃 | barplot | 성별 생존율 비교 |
| 여러 변수 | pairplot | 주요 변수 관계 한 번에 확인 |

확인 질문:

- 타깃 값에 따라 분포가 달라지는 변수가 있는가?
- 두 수치형 변수가 같이 증가하거나 감소하는가?



## 07. 상관관계 및 교차분석

변수 간 관계를 숫자와 표로 확인.

수치형 변수:

- 상관계수 확인
- heatmap으로 시각화
- `1`에 가까우면 양의 관계
- `-1`에 가까우면 음의 관계
- `0`에 가까우면 선형 관계가 약함

범주형 변수:

- 교차표 확인

주의할 점:

정리할 내용:

- 인코딩이 필요한 범주형 컬럼
- 스케일링이 필요한 수치형 컬럼

중요한 점:

> EDA 단계에서는 무조건 데이터를 고치는 것이 아니라, 왜 고쳐야 하는지 근거를 정리해야 함.



```python
# 원본 보존 후 복사본에서 전처리
df_clean = df.copy()

# 결측치 대체 예시
df_clean["Age"] = df_clean["Age"].fillna(df_clean["Age"].median())

# 불필요한 컬럼 제거 예시
df_clean = df_clean.drop(columns=["Cabin"])
```



## 09. Feature Engineering 방향 정리

Feature Engineering: 기존 컬럼을 이용해 분석이나 모델링에 더 도움이 되는 새 변수를 만드는 과정.

예시:

| 원본 컬럼 | 파생 변수 예시 | 의미 |
|---|---|---|
| 나이 | 나이 그룹 | 어린이, 청년, 성인 등으로 구간화 |
| 형제/배우자 수 + 부모/자녀 수 | 가족 규모 | 함께 탑승한 가족 수 |
| 가족 규모 | 혼자 탑승 여부 | 단독 탑승 여부 |
| 요금 | 로그 요금 | 치우친 요금 분포 완화 |
| 객실 번호 | 객실 정보 유무 | 객실 정보가 있는지 여부 |

확인 질문:

- 기존 컬럼 조합으로 더 의미 있는 변수를 만들 수 있는가?
- 수치형 변수를 구간화하면 해석이 쉬워지는가?
- 값의 치우침을 줄이기 위한 변환이 필요한가?


```python
# 가족 규모 파생변수
df["FamilySize"] = df["SibSp"] + df["Parch"] + 1

# 혼자 탑승 여부 파생변수
df["IsAlone"] = (df["FamilySize"] == 1).astype(int)

# 나이 구간 파생변수
df["AgeGroup"] = pd.cut(
    df["Age"],
    bins=[0, 12, 18, 35, 60, 100],
    labels=["Child", "Teen", "YoungAdult", "Adult", "Senior"]
)
```



## 10. 최종 요약 및 인사이트 도출

EDA의 마지막은 그래프가 아니라 해석 문장임.

정리할 내용:

1. 데이터 구조 요약
2. 주요 결측치와 품질 이슈
3. 중요한 분포 특징
4. 타깃과 관련 있어 보이는 변수
5. 전처리 필요 사항
6. Feature Engineering 아이디어
7. 다음 분석 또는 머신러닝으로 이어질 액션

좋은 인사이트 문장 예시:

- 전체 생존율은 약 38%로 사망자 비율이 더 높음.
- 여성의 생존율이 남성보다 높게 나타남.
- 객실 등급이 높을수록 생존율이 높은 경향이 있음.
- `Age`, `Cabin`, `Embarked` 컬럼에는 결측치가 있어 처리 전략이 필요함.
- `FamilySize`, `IsAlone`, `HasCabin`, `LogFare` 같은 파생변수를 만들 수 있음.

간단한 힌트 코드:

```python
# 전체 생존율
survival_rate = df["Survived"].mean()

# 성별 생존율
sex_survival = df.groupby("Sex")["Survived"].mean()

# 객실 등급별 생존율
pclass_survival = df.groupby("Pclass")["Survived"].mean()

print("전체 생존율:", survival_rate)
print("성별 생존율:")
print(sex_survival)
print("객실 등급별 생존율:")
print(pclass_survival)
```
